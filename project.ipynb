```python
import pandas as pd

#mysql toolkit to connect with mysql and setup everything
import pymysql # this will work as adapter
from sqlalchemy import create_engine
```

```python
print(pd.__version__)
```

```python
# encoding_errors will ignore all the errors or unformated exceptions that will cause of pain in the data set.
df = pd.read_csv(r"C:\Users\Pranay Joshi\Desktop\Project Walmart\Walmart.csv", encoding_errors='ignore')

print(f'Total no. of rows and columns in the dataset :- {df.shape}')
```

```python
df.head()
```

```python
# Statistics of the dataset's numerical columns

df.describe()
```

```python
#data info of the dataset

df.info()

# This column 'unit_price' need to convert into numerical col and for 
# both col 'unit_price & quantity' we need to check the n/a values and fill 
# then with either median or we can remove them also.
```

```python
# all duplicate values count in the data set.

df.duplicated().sum()
```

```python
# To simply remove the duplicates we use 'drop_duplicates(inplace = True)

df.drop_duplicates(inplace = True)
```

```python
# find if there any missing values or not.

df.isnull().sum()

# it will return all the count of missing values in the columns
```

```python
# Treating missing values of the both columns by deleting the rows simply, 
# as we don't even given the price as well so it will create the issue in future, 
# that's why instead of replacing the null values with midean we are going to delete the null values.

df.dropna(inplace = True)

# Verify 

df.isnull().sum()
```

```python
df.shape
```

```python
# We needed to convert the unit_price to int 

df.dtypes
```

```python
# first we need to remove the '$' from the values
#  because the python is not able to identify is the value is numeric or not,
# till we have '$' in the value and shows it as 'String'.

df['unit_price'] = df['unit_price'].str.replace('$','').astype(float)

df.head()
```

```python
# verify 

df.info()
```

```python
df.columns

#all the columns in the data set
```

```python
# Here we have created the new column 'Total_Amt' which shows the total amount of sold per invoice

df['Total_Amt'] = round(df['unit_price']*df['quantity'],2)

df.head(2)
```

```python
# saving the file in .csv format

df.to_csv('Walmart_Final_data', index=False)
```
